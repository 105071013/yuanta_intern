```python
def create_assist_date(datestart = None,dateend = None):
# 创建日期辅助表

    if datestart is None:
        datestart = '20190812'
    if dateend is None:
        dateend = '20200319'

    # 转为日期格式
    datestart=datetime.datetime.strptime(datestart,'%Y%m%d')
    dateend=datetime.datetime.strptime(dateend,'%Y%m%d')
    date_list = []
    date_list.append(datestart.strftime('%Y%m%d'))
    while datestart<dateend:
        # 日期叠加一天
        datestart+=datetime.timedelta(days=+1)
        # 日期转字符串存入列表
        date_list.append(datestart.strftime('%Y%m%d'))
        

    return date_list
```

```python
def createdate(datestart=None ,dateend =None):


    if datestart is None:
        datestart = '2020-01-21 02:00:00.000'
    if dateend is None:
        dateend = '2019-12-31 14:50:40.851'

  
    datestart=datetime.datetime.strptime(datestart,'%Y-%m-%d %H:%M:%S.%f')
    dateend=datetime.datetime.strptime(dateend,'%Y-%m-%d %H:%M:%S.%f')
    date_list = []
    date_list.append(datestart.strftime('%Y-%m-%d %H:%M:%S.%f'))
    while datestart>dateend:
        datestart+=datetime.timedelta(minutes=-5)
        date_list.append(datestart.strftime('%Y-%m-%d %H:%M:%S.%f'))

    return date_list
```

```python
def createdate1(datestart=None ,dateend =None):


    if datestart is None:
        datestart = '2020-01-01 14:50:40.851'
    if dateend is None:
        dateend = '2020-03-21 09:00:00.000'

   
    datestart=datetime.datetime.strptime(datestart,'%Y-%m-%d %H:%M:%S.%f')
    dateend=datetime.datetime.strptime(dateend,'%Y-%m-%d %H:%M:%S.%f')
    date_list = []
    date_list.append(datestart.strftime('%Y-%m-%d %H:%M:%S.%f'))
    while datestart<dateend:
        datestart+=datetime.timedelta(minutes=+5)
        date_list.append(datestart.strftime('%Y-%m-%d %H:%M:%S.%f'))

    return date_list
```

```python
def label_y(x):
    if ((x['std']*2)<= x['ret']):
        x['y'] = 1 
    elif ((x['std']*(-2))>= x['ret']):
        x['y'] = -1
    elif  (((x['std']*(-2))< x['ret'])|((x['std']*(2))> x['ret'])):
        x['y'] = 0
    return x
```
